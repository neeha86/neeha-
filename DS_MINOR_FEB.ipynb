**6. How many different vehicles are present in this data ?**

```python
data = pd.read_csv("https://raw.githubusercontent.com/ameer-aiml/ai-vodafone/main/datasets/car_data.csv")
df=pd.DataFrame(data)
cnt=0
visited=[]
for i in range(0,len(df['Car_Name'])):
  if df['Car_Name'][i] not in visited:
    visited.append(df['Car_Name'][i])
    cnt += 1
print("no.of.different vehicles present in this data : ",cnt)
print("unique vehicle names : ",visited)
```

**12. Which is the most and least cost depreciated vehicle in data
?**

```python
data = pd.read_csv("https://raw.githubusercontent.com/ameer-aiml/ai-vodafone/main/datasets/car_data.csv")
df=pd.DataFrame(data)
df['depreciated'] = df['Present_Price']-df['Selling_Price']
print("least cost depreciated vehicle ")
df[df.depreciated == df.depreciated.min()]
```

```python
print("most cost depreciated vehicle ")
df[df.depreciated == df.depreciated.max()]
```
